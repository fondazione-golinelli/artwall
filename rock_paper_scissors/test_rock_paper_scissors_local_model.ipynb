```python
def classes_to_label(c):
    if c[0][0]:
        return "PAPER"
    elif c[0][1]:
        return "ROCK"
    elif c[0][2]:
        return "SCISSORS"
    return "UNKNOWN"
```

```python
import os
import tensorflow as tf


model_path = os.path.join(os.getcwd(), 'rps.h5')
model = tf.keras.models.load_model("rps.h5")

model.summary()
```

```python
challenge_directory = os.path.join(os.getcwd(), 'images', 'rps-challenge')
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from keras.preprocessing import image


for filename in os.listdir(challenge_directory):
    
    # predicting images
    path = os.path.join(challenge_directory, filename)
    img = image.load_img(path, target_size=(150, 150))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)

    images = np.vstack([x])
    classes = model.predict(images, batch_size=10)
    img = mpimg.imread(path)
    plt.imshow(img)
    plt.axis('Off')
    plt.show()
    print(filename)
    print(classes)
    print(classes_to_label(classes))
```
